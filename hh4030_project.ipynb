```python
import folium
m=folium.Map(location=[1.3413,103.9245])
m
```

```python
from branca.element import Figure
fig=Figure(width=550,height=350)
```

```python
m1=folium.Map(width=550,height=350,location=[1.3413,103.9245], zoom_start=40,min_zoom=40,max_zoom=150)
fig.add_child(m1)
m1
```

```python
fig2=Figure(width=550,height=350)
m2=folium.Map(width=550,height=350,location=[1.3413,103.9245])
fig2.add_child(m2)

# Add Tile Layers with attribution - focusing on colorful styles
folium.TileLayer(
    tiles='Stamen Watercolor',  # Prioritize Stamen Watercolor for color
    attr='Map tiles by <a href="http://stamen.com">Stamen Design</a>, <a href="http://creativecommons.org/licenses/by/3.0">CC BY 3.0</a> &mdash; Map data &copy; <a href="http://www.openstreetmap.org/copyright">OpenStreetMap</a> contributors',
).add_to(m2)

# Include other tile options for variety
folium.TileLayer(
    tiles='Stamen Terrain',
    attr='Map tiles by <a href="http://stamen.com">Stamen Design</a>, <a href="http://creativecommons.org/licenses/by/3.0">CC BY 3.0</a> &mdash; Map data &copy; <a href="http://www.openstreetmap.org/copyright">OpenStreetMap</a> contributors',
).add_to(m2)

folium.TileLayer(
    tiles='Stamen Toner',
    attr='Map tiles by <a href="http://stamen.com">Stamen Design</a>, <a href="http://creativecommons.org/licenses/by/3.0">CC BY 3.0</a> &mdash; Map data &copy; <a href="http://www.openstreetmap.org/copyright">OpenStreetMap</a> contributors',
).add_to(m2)
#might not need the first 3 titles?

folium.TileLayer(
    tiles='cartodbpositron',
    attr='CartoDB',
).add_to(m2)

folium.TileLayer(
    tiles='OpenStreetMap',
    attr='&copy; <a href="http://www.openstreetmap.org/copyright">OpenStreetMap</a> contributors'
).add_to(m2)


folium.LayerControl().add_to(m2)
m2
```

```python
# creating basemap
fig3=Figure(width=500,height=400)
m3=folium.Map(location=[1.3413,103.9245],tiles='cartodbpositron',zoom_start=11)
fig3.add_child(m3)

# adding markers to the map
folium.Marker(location=[1.346274,103.921973],popup='Tree Species 1',tooltip='<strong>Click here to see Pop-up</strong>').add_to(m3)
folium.Marker(location=[1.339939,103.920955],popup='<strong>Marker3</strong>',tooltip='<strong>Click here to see Popup</strong>').add_to(m3)
folium.Marker(location=[1.345005,103.929024],popup='<h3 style="color:green;">Marker2</h3>',tooltip='<strong>Click here to see Popup</strong>').add_to(m3)

m3
```

```python
# Creating Basemap
fig4=Figure(height=350,width=550)
m4=folium.Map(location=[28.4911091,77.0867361],tiles='cartodbpositron',zoom_start=11)
fig4.add_child(m4)

# Adding Custom Markers
folium.Marker(location=[28.4211091,77.0267361],popup='Custom Marker 1',tooltip='<strong>Click here to see Popup</strong>',icon=folium.Icon(color='red',icon='none')).add_to(m4)
folium.Marker(location=[28.4411091,77.1167361],popup='Custom Marker 2',tooltip='<strong>Click here to see Popup</strong>',icon=folium.Icon(color='green',prefix='glyphicon',icon='off')).add_to(m4)
folium.Marker(location=[28.4511091,77.1667361],popup='Custom Marker 3',tooltip='<strong>Click here to see Popup</strong>',icon=folium.Icon(color='purple',prefix='fa',icon='anchor')).add_to(m4)

m4
```
